```python
%load_ext autoreload
%autoreload 2

import pandas as pd

from sklearn.model_selection import train_test_split
```

```python
data = pd.read_excel("./data/OCR_verified.xlsx")
```

```python
data_clean = data[["Erro", "SEC", "TEXTO BLOCO", "Nº LINHAS TOTAIS"]]; data_clean.head()
```

```python
data_unsup = data_clean[data_clean["Erro"] !=1][["TEXTO BLOCO"]]; data_unsup, len(data_unsup), len(data_clean)
```

```python
X = data_clean[data_clean["Erro"] == 1][["SEC", "TEXTO BLOCO"]]; X
```

```python
X = X.groupby("SEC").filter(lambda x: len(x) > 3)
```

```python
X["SEC"].unique(); len(X["SEC"].unique())
```

```python
X, y = X["TEXTO BLOCO"].to_list(), X["SEC"].to_list()
```

```python
len(X), len(y), X[0], y[0]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42, stratify=y)
```

```python
train_csv = pd.DataFrame()
test_csv = pd.DataFrame()
```

```python
train_csv["label"], train_csv["text"] = y_train, X_train
test_csv["label"], test_csv["text"] = y_test, X_test
```
